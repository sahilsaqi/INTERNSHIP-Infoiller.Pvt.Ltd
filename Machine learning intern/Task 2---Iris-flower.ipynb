## Step 1 - Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("iris.csv")
#display first 5 rows from table
df.head()
```

## Building Machine Learning Models

```python
# Split the data into train and test data
X=df[['sepal length','sepal width','petal length','petal width']]
y=df['class']

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=10)
```

### 1. LOGISTIC REGRESSION

```python
lreg=LogisticRegression()
lreg.fit(X_train,y_train)
y_pred=lreg.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy Score of Logistic Regression is :",accuracy_score(y_pred,y_test))
```

### 2. SVM(Support Vector Machine)

```python
svm=SVC()
svm.fit(X_train,y_train)
y_pred=svm.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy Score of SVM is :",accuracy_score(y_pred,y_test))
```

### 3. DECISION TREES

```python
dtreg=DecisionTreeClassifier()
dtreg.fit(X_train,y_train)
y_pred=dtreg.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy Score of Decision Tree is :",accuracy_score(y_pred,y_test))
```

### 4. NAIVE BAYES CLASSIFIER

```python
nbreg=GaussianNB()
nbreg.fit(X_train,y_train)
y_pred=nbreg.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy Score of Naive Bayes classifcation is :",accuracy_score(y_pred,y_test))
```

### 5. K-NEAREST NEIGHBOUR

```python
kreg=KNeighborsClassifier(n_neighbors=6)
kreg.fit(X_train,y_train)
y_pred=kreg.predict(X_test)
print(classification_report(y_test,y_pred))
print("Accuracy Score of K-Nearest Neighbor is :",accuracy_score(y_pred,y_test))
```
